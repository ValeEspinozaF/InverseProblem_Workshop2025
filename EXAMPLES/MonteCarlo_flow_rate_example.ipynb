```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Define the mean values and standard deviations for the variables
r_mean, r_std = 0.01, 0.001         # Radius in meters
P_mean, P_std = 1000, 100           # Pressure difference in Pascals
mu_mean, mu_std = 0.001, 0.0001     # Viscosity in PaÂ·s
L_mean, L_std = 1.0, 0.1            # Length in meters

# Number of Monte Carlo samples
n_samples = 10000

# Generate random samples for each variable with Gaussian distribution
r = np.random.normal(r_mean, r_std, n_samples)
P = np.random.normal(P_mean, P_std, n_samples)
mu = np.random.normal(mu_mean, mu_std, n_samples)
L = np.random.normal(L_mean, L_std, n_samples)

# Compute Q using the Darcy-Weisbach equation
Q = (np.pi * r**4 * P) / (8 * mu * L)
```

```python
# Plot the histogram of Q
fig, ax = plt.subplots(figsize=(7,4))
sns.histplot(Q, bins=50, ax=ax)

# Plot mean value
ymin, ymax = ax.get_ylim()
ax.plot([np.mean(Q), np.mean(Q)], [0, ymax*1.2], color='red', linestyle='--', label='Average Q')

# Polish the plot
ax.legend()
ax.set(xlabel='Volumetric Flow Rate (Q)', ylabel='Frequency', title='Distribution of Q', ylim=(0, ymax*1.2));
```
